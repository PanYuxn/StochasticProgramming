基本决策存在两种：

1. 第一阶段需要被确定的决策，这部分决策不存在不确定性。

2. 第二阶段决策需要进行数值实验才可知具体结果。



下面举一个书本当中的例子来看一下不同阶段的考虑：

一个旅行推销员每天收到一件商品，她拜访客户希望出售该物品。当找到买家或拜访所有客户时，她就回家了。客户以随机的方式购买或不购买。决策不受前几天决策的影响。销售人员希望确定拜访客户的先后顺序，尽可能早到家。花费的时间包括在每个被访问客户处的旅行时间加上一些服务时间。



上面这个问题存在两个阶段的决策，第一个阶段是：选择固定的访问顺序，第二个阶段是：当前顾客选择购买与否对后续的影响。在这个情况下，具体的答案如下：

1. 如果选择访问顺序是(1,2)，从8：00出发，那么9：30到达客户1，此时如果客户购买可以直接回去，到达家里时间为11：00；如果客户不购买，那么需要前往客户2在回家，最终到达时间为16：30。
2. 如果选择访问顺序是(2,1)，从8：00出发，那么12到达客户2，此时如果客户购买可以之恶回去，到达家里时间为16：00；如果客户不购买，那么需要前往客户1再回家，最终到达时间为16：30。

上面这个解决方案其实很明显，访问顺序(1,2)才是比较合理的选择。
