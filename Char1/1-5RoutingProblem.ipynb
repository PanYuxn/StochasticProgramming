# Routing Problem

这是一个简单的取送货问题，但是其中的某个deliver存在一个随机值，他可能有两种production，图片如下。



## Wait-and-see Solution

第一个问题类型是我提前知道了具体的生产信息，也就是我会等到我知道该deliver的具体demand的情况下才进行决策。这种情况下的答案是：$$WS=\frac{1}{2}10+\frac{1}{2}14=12$$


## Expected Value Solution

第二种解决方法是把不确定性当作期望来考虑，也就是$\frac{1+7}{2}=4$，然后进行决策。在这种情况下，我们的决策其实是一次性走完所有路程.但是需要考虑的是，如果deliver的demand是7，那么我们可能需要进行一次折返，这样就会导致cost增加。这种情况下的最优值为：$$EEV=\frac{1}{2}10+\frac{1}{2}18=14$$




## Recourse Solution

第三种解决方案是我们去正常对待不确定性，然后再面对不确定性的时候尽可能做出一些`补偿措施`来减少我们的损失。这里的不确定性只有deliver C的demand，因此我们如果第一个访问deliver C，那么我们可以得到具体的不确定值。最终的最优值为：$$RP=\frac{1}{2}11+\frac{1}{2}17=12.5$$



## 分析

* EVPI = RP-WS = 12.5 -12 = 0.5 
在Perfect information的情况下，我们会有0.5的收益。

* VSS = EEV - RP = 14 - 12.5 = 1.5
如果我们尝试把不确定性忽视，在这个情况下，我们会有1.5的损失存在。


## 更常见的情况

上面的例子可能比较极端，更可能的情况是这个demand是一个符合均匀分布的函数，它的均值是4，在1-8之间震荡。在这种情况下，我们需要对最终的Expected Cost进行修改，也就是我们需要增加上对应的不同可能性。

## Chance Constraint

还有一种情况是Chance Constraint，也就是我们需要95%的情况下可以满足我们的需求。在这种情况下，我们需要对我们的决策进行修改。（这个感觉算是后续的扩展了）
